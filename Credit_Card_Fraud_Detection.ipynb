```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('creditcard[1].csv')
data.head()
```

```python
# data set info
data.info()
```

```python
# checking th no of missing values
data.isnull().sum()
```

```python
# distribution of legit tranction & fradulent transaction
data['Class'].value_counts()
```

The data set is highly unbalanced

1-normal transaction
2-fraudulent transaction

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
#statistical measure
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing the values
data.groupby('Class').mean()
```

```python
print(legit.shape)
print(fraud.shape)
```

buid a sample dataset Containing similar distribution of normal transaction and fraudulent transaction

```python
legit_sample = legit.sample(n=492)
```

Concatenating 2 dudder frames

```python
new_data = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
x = new_data.drop(columns='Class', axis=1)
y = new_data['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

Model evaluation

Accuracy Score

```python
from sklearn.metrics import accuracy_score
x_predict = model.predict(x_train)
print("Training Data accuracy",accuracy_score(x_predict, y_train))
```

```python
# accuracy on test data
x_test_predict = model.predict(x_test)
print("Test Data accuracy",accuracy_score(x_test_predict, y_test))
```
